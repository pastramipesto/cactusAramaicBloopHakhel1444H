```python
#while loop = execute some code WHILE some condition remains true
#first example
#notice the reprompt is not infinite
# prompt being asked twice, it kept asking until you answer
name = input("Enter your name: ")

while name == "":
  print("You did not enter your name")
  name = input("Enter your name: ")

print(f"Hello {name}")
```

```python
#second example, becomes infinite, without asking input, it repeats itself,
#... when you hit enter. NOTICE, no Input prompt. Works fine
# prompt being asked once but it gets broken in the head without the second rep
name = input("Enter your name: ")

while name == "":
  print("You did not enter your name")

print(f"Hello {name}")
```

```python
#example 3
# it keeps asking only when negative

age = int(input("Enter your age: "))

while age <0:
  print("Age cannot be negative.")
  age = int(input("Enter your age: "))

print(f"You are {age} years old")
```

```python
#example 4
#this one will keep on talking ans asking you questions until you hit q for quit

food = input("Enter a food you like (q to quit): ")

while not food == "q":
  print(f"You like {food}")
  food = input("Enter another food you like (q to quit): ")

print("bye")
```

```python
#example 5
#using or to filter, a range.

num = int(input("Enter a number between 1-10: "))

while num < 1 or num > 10:
  print(f"{num} is not valid")
  num = int(input("Enter a number between 1-10: "))

print(f"Your number is {num}")
```
